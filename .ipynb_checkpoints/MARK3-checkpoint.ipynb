```python
# Visualizations
import matplotlib.pyplot as plt

from scipy.stats import poisson
```

```python
# Mark3 Mean
def Mark3Mean(n90, n50, n10, n05, n01):
    return 0.233 * n90 + 0.4 * n50 + 0.255 * n10 + 0.045 * n05 + 0.03 * n01

#S
def S(n90, n50, n10, n05, n01):
    return 0.121 - 0.237 * n90 - 0.093 * n50 + 0.183 * n10 + 0.073 * n05 + 0.123 * n01
```

## Números S

```python
# Los n son número de yacimientos

def s1(n90):
    return 1 + 2 * n90

def s2(n90,n50):
    if n50 == n90:
        result = 1
    elif n50 > n90:
        result = 2 + 2 * ( n50 - n90 - 1)
    
    return result

def s3(n50, n10):
    if n50 == n10:
        result = 1
    elif n10 > n50:
        result = 2 + 2 * ( n10 - n50 - 1)
    
    return result
```

## Probabilidades p(n)

Los percentiles 90, 50, 10 pueden representar un modelo de distribución por defecto.

Si no se usa la districión por defecto se selecciona la distribución de los modelos de depósito de dos formas:

* Aproximando una distribución lognormal para grados y tonelajes
*  approximating piecewise linear distributions (empirical option)

```python
def depProb(nYac,n90,n50,n10):
    # .1 .4 son las probabilidades de p90, p10 y alredeor de p50
    
    ns1 = s1(n90)
    ns2 = s2(n90, n50)
    ns3 = s3(n50, n10)
    
    #Caso I
    if n90 < n50 < n10:
        if nYac < n90:
            yacProb = 2 * ( .1 / ns1 ) 
        elif nYac == n90:
            yacProb = ( .1 / ns1 ) + ( .4 / ns2 )
        elif n90 < nYac < n50:
            yacProb = 2 * ( .4 / ns2 )
        elif nYac == n50:
            yacProb = (.4 / ns2) + (.4 / ns3)
        elif n50 < nYac < n10:
            yacProb = 2 * (.4 / ns3)
        elif nYac == n10:
            yacProb = ( .4 / ns3) + .1
            #yacProb = 0.08
        elif nYac > n10:
            yacProb = 0
           
    #Caso II
    if n90 == n50 < n10:
        if nyac < n90:
            yacProb = 2 * ( .1 / ns1 )
        elif nyac == n90:
            yacProb = ( .1 / ns1 ) + .4 + ( .4 / ns3 )
        elif n50 < nYac < n10:
            yacProb = 2 * ( .4 / ns3 )
        elif nYac == n10:
            yacProb = ( .4 / ns3 ) + .1
        elif nYac > n10:
            yacProb = 0
            
    #Caso III
    if n90 < n50 == n10:
        if nYac < n90:
            yacProb = 2 * ( .1 / ns1 )
        elif nYac == n90:
            yacProb = ( .1 / ns1 ) + ( .4 / ns2 )
        elif n90 < nYac < n50:
            yacProb = 2 * ( .4 / ns2 )
        elif nYac == n50:
            yacProb = ( .4 / ns2) + ( .4 / ns3 ) + .1
        elif nYac > n10:
            yacProb = 0
            
    #Caso IV
    if n90 == n50 == n10:
        if nYac < n90:
            yacProb = 2 * ( .1 / ns1)
        elif nYac == n90:
            yacProb = ( .1 / ns1 ) + ( .4 / ns2 ) + ( .4 / ns3 ) + .1
        elif nYac > n10:
            yacProb = 0
    
            
    return yacProb
```

```python
def Mark3Plot(depNum, Prob):
    fig, ax = plt.subplots(figsize=(16,8))
    ax.plot(depNum, Prob, marker='o');

    # tidy up the figure
    ax.grid(True)

    ax.set_title('Cálculo del Número de Depósitos con el Algoritmo MARK3')
    ax.set_xlabel('Número de Depósitos')
    ax.set_ylabel('Probabilidad')

    #plt.text(0,0, f'Número de depósitos esperado: {depMean:.1f}')
    plt.show()
    
def Mark3PlotSeveral(depNum, yLst):
    fig, ax = plt.subplots(figsize=(16,8))
    
    for y in yLst:
        ax.plot(depNum, y, marker='o');

    # tidy up the figure
    ax.grid(True)

    ax.set_title('Cálculo del Número de Depósitos con el Algoritmo MARK3')
    ax.set_xlabel('Número de Depósitos')
    ax.set_ylabel('Probabilidad')

    #plt.text(0,0, f'Número de depósitos esperado: {depMean:.1f}')
    plt.show()
```

## Empirical Distributions

An empirical distribution function provides a way to model and sample cumulative probabilities for a data sample that does not fit a standard probability distribution.

As such, it is sometimes called the empirical cumulative distribution function, or ECDF for short.
An empirical probability density function can be fit and used for a data sampling using a nonparametric density estimation method, such as Kernel Density Estimation (KDE).

An empirical cumulative distribution function is called the Empirical Distribution Function, or EDF for short. It is also referred to as the Empirical Cumulative Distribution Function, or ECDF.

The EDF is calculated by ordering all of the unique observations in the data sample and calculating the cumulative probability for each as the number of observations less than or equal to a given observation divided by the total number of observations.

As follows:

    EDF(x) = number of observations <= x / n
